## SVM para Regresión

```python
# Importamos las librerías necesarias.
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ==========================================
# DATOS GENERADOS DE MANERA ALEATORIA
# ==========================================
X = np.sort(5*np.random.rand(200,1),axis=0)
Y = np.sin(X).ravel()
Y[::5] += 3*(0.5 - np.random.rand(40))
```

```python
# ==========================================
# GRÁFICO
# ==========================================
plt.scatter(X,Y, color="darkorange", label="data")
plt.legend()
```

```python
# ==========================================
# MODELOS
# ==========================================
from sklearn.svm import SVR

# Usaremos distintos kernel para ver cuál se ajusta mejor a los datos
C=1e3
svr_lin = SVR(kernel="linear", C=C)
svr_rbf = SVR(kernel="rbf", C=C, gamma=0.1)
svr_pol = SVR(kernel="poly", C=C, degree=3)
```

```python
# ==========================================
# PREDICCIONES
# ==========================================
y_lin = svr_lin.fit(X,Y).predict(X)
y_rbf = svr_rbf.fit(X,Y).predict(X)
y_pol = svr_pol.fit(X,Y).predict(X)
```

```python
# ==========================================
# GRÁFICO DE LOS MODELOS
# ==========================================

lw = 4
plt.figure(figsize=(16,9))
plt.scatter(X,Y,color="darkorange", label ="data")
plt.plot(X,y_lin, color="navy", lw = lw, label = "SVM Lineal")
plt.plot(X,y_rbf, color="c", lw=lw, label="SVM Radial")
plt.plot(X,y_pol, color="cornflowerblue", lw=lw, label="SVM Polinómico")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Support Vector Regression")
plt.legend()
plt.show()
```

```python
# ==========================================
# ERROR DEL MODELO
# ==========================================

from sklearn.metrics import mean_squared_error

mse_lin = mean_squared_error(Y, y_lin, squared=False)
mse_rbf = mean_squared_error(Y, y_rbf, squared=False)
mse_pol = mean_squared_error(Y, y_pol, squared=False)

mse =[mse_lin, mse_rbf, mse_pol]
mse
```

El SVM Radial es el que tiene menor error.
